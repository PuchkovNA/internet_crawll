```python
category_and_subcategory_list=[]
items=[]
for category in product_category.ul:
        try:
            if isinstance(category, Tag):
                 category_and_subcategory_list.append([{category.a.text:category.a['href']},  dict(zip([item.a.text for item in category.find_all('li')],  [item.a['href'] for item in category.find_all('li')]))])
        except NavigableString: 
            pass
```

```python
list_of_tupile_of_products_by_all_categories_and_subcategories=[]
for category_and_subcategory_item in category_and_subcategory_list:
    print(next(iter(category_and_subcategory_item[0])))
    category=next(iter(category_and_subcategory_item[0]))
    for sub_category in category_and_subcategory_item[1].keys():
        print(sub_category)
        for j in range(1,30):
            print(f'page {j}')
            url=category_and_subcategory_item[1].get(sub_category)+'?sf_paged='+str(j)+'&sort_order=_sfm_product_rating+desc+num&_sfm_quality_label_filter=verified'
            headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"}
            req=requests.get(url, headers=headers)
            soup = BeautifulSoup(req.text, "lxml")
            print(url)
            # print(soup.find('ul', {'class':'products columns-4'}))
            if soup.find('ul', {'class':'products columns-4'}):
                for item in soup.find('ul', {'class':'products columns-4'}):
                    try:
                        if isinstance(item, Tag):
                            if isinstance(item, Tag):
                                try:
                                    product_title=item.find('h2', {'class':'woocommerce-loop-product__title'}).text
                                except AttributeError:
                                    continue 

                            if isinstance(item, Tag):
                                try:
                                    _id=item.find('div', {'class':'tinv-wraper woocommerce tinv-wishlist tinvwl-shortcode-add-to-cart tinvwl-loop-button-wrapper'})['data-tinvwl_product_id']
                                except AttributeError:
                                    continue 
                            if isinstance(item, Tag):
                                try:    
                                    brand=item.find('p', {'class':'brand-text'}).text
                                except AttributeError:
                                    continue 
                            if isinstance(item, Tag):
                                try:
                                    product_link=item.find('a', {'class':'woocommerce-loop-product__link'})['href']
                                except AttributeError:
                                    continue
                            if isinstance(item, Tag):
                                try:
                                    rating=item.find('div', {'class':'rating-figure'}).span.text
                                except AttributeError:
                                    continue
                            if isinstance(item, Tag):
                                try:
                                    quality=item.find('span', string='Качество').previous_sibling.previous_sibling.span.text
                                except AttributeError:
                                    continue    

                        myTuple=( {"_id":_id,
                                  "product_title":product_title,
                                  "product_link":product_link,
                                  "brand":brand,
                                  "rating":rating,
                                  "quality":quality,
                                  "sub_category":sub_category,
                                   "category":category} )
                        list_of_tupile_of_products_by_all_categories_and_subcategories.append(myTuple)
                    except NavigableString:     
                        pass
                    # ops.append(item)
                    # print(item)
            else:
                break
```

```python
from pymongo import UpdateOne
def insert_or_update_products_in_all_categories(collection_name, list_name):
    upserts=[ UpdateOne({'_id':x['_id']}, {'$setOnInsert':x}, upsert=True) for x in list_name]
    collection_name.bulk_write(upserts)


```

```python
insert_or_update_products_in_all_categories(collection_name, list_of_tupile_of_products_by_all_categories_and_subcategories)
#collection_name.drop()
```

```python
def select_products_with_rating_and_quality_greater_then(rating_and_quality):
    for item in  collection_name.find({ 'rating': {'$gt':rating_and_quality }, 'quality': {'$gt':rating_and_quality }}):
        print(item)
```
